```python
import json
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
```

```python
archivo = "planta0recorridos.json"
with open(archivo, "r") as file:
    recorridos0 = json.load(file)

archivo = "planta1recorridos.json"
with open(archivo, "r") as file:
    recorridos1 = json.load(file)

archivo = "planta2recorridos.json"
with open(archivo, "r") as file:
    recorridos2 = json.load(file)

recorridos = [recorridos0, recorridos1, recorridos2]
```

```python

for recorrido in recorridos:
    # Queremos ver la distribucion de los recorridos
    # Para eso, vamos a graficar un histograma de los recorridos

    # Primero, vamos a ver cuantos recorridos hay en total
    # Para eso, vamos a contar cuantos elementos hay en cada lista
    # y luego sumarlos
    cantidad_recorridos = 0
    for dia in recorrido.values():
        cantidad_recorridos += len(dia)

    # Ahora, vamos a crear una lista con todos los recorridos
    # Para eso, vamos a recorrer cada dia, y agregar cada recorrido
    # a la lista
    lista_recorridos = []
    for dia in recorrido.values():
        for r in dia:
            lista_recorridos.append(r)

    # Creamos un mapa de calor con los recorridos
    # Para eso, vamos a crear una matriz de 10x10
    # y vamos a ir sumando 1 en cada celda por cada recorrido
    # que pase por esa celda
    mapa_calor = np.zeros((10, 10))

    # sumamos 1 a cada celda de la matriz
    for x in range(10):
        for y in range(10):
            mapa_calor[x][y] += 1

    for r in lista_recorridos:
        x = int(r[0])
        y = int(r[1])
        mapa_calor[x][y] = mapa_calor[x][y] * 1.01

    # Ahora, vamos a graficar el mapa de calor con colores de verde a rojo
    fig, ax = plt.subplots()
    cmap = mcolors.ListedColormap(['green', 'yellow', 'orange', 'red'])
    im = ax.imshow(mapa_calor, interpolation="nearest", cmap='summer')
    fig.colorbar(im)
    plt.show()

    # imprimir la cantidad de recorridos
    print("Cantidad de recorridos: ", cantidad_recorridos)
```
